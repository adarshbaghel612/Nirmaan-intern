```python
with open('/content/Sample text for case study.txt', 'r') as file:
    text = file.read()

print(text)
```

```python
Salutaions=["Hii","Hello,","Good Morning","Good Afternoon", "Good Evening", "Good Day", "Hello everyone", "I am excited to introduce","Feeling great"]
text_lower = text.lower()

found_words = [word for word in Salutaions if word.lower() in text_lower]

print("Found:", found_words)
```

```python
Score=0
Normal=["Hii","Hello,"]
Good=["Good Morning","Good Afternoon", "Good Evening", "Good Day", "Hello everyone"]
Excellent=["I am excited to introduce","Feeling great"]
```

```python
for i in range(len(Normal)):
  if Normal[i] in found_words:
    Score+=2
for i in range(len(Good)):
  if Good[i] in found_words:
    Score+=4
for i in range(len(Excellent)):
  if Excellent[i] in found_words:
    Score+=5
```

```python
print(Score)
```

```python
import re

features = {}

if isinstance(text, set):
    text = list(text)[0]

# Extract Name
name_match = re.search(r"(myself|i am|this is)\s+([A-Za-z]+)", text, re.IGNORECASE)
if name_match:
    features["name"] = name_match.group(2)
    Score+=4
else:
  features["name"] = "Not Found"

# Extract Age
age_match = re.search(r"(\d{1,2})\s+years?\s+old", text)
if age_match:
    features["age"] = age_match.group(1)
    Score+=4
else:
  features["age"] = "Not Found"

# Extract Class
class_match = re.search(r"class\s+([\w\s]+)", text, re.IGNORECASE)
if class_match:
    features["class"] = class_match.group(1).strip().split(" ")[0]  # Pick only "8th"
    Score+=4
else:
  features["class"] = "Not Found"

# Extract School Name
school_match = re.search(r"from\s+([A-Za-z\s]+School)", text, re.IGNORECASE)
if school_match:
    features["school"] = school_match.group(1).strip()
    Score+=4
else:
  features["school"] = "Not Found"

hobby_keywords = ["cricket", "football", "reading", "drawing", "dance", "music"]
found_hobbies = [word for word in hobby_keywords if word.lower() in text.lower()]
features["hobbies"] = list(set(found_hobbies))
Score+=4

    # Favorite Subject
subject_match = re.search(r"favorite subject is ([A-Za-z]+)", text, re.IGNORECASE)
if subject_match:
    features["favorite_subject"] = subject_match.group(1)
else:
  features["favorite_subject"] = "Not Found"

    # Fun Fact Extraction
fun_fact_match = re.search(r"fun fact about me is that (.+?)(\.|$)", text, re.IGNORECASE)
if fun_fact_match:
    features["fun_fact"] = fun_fact_match.group(1)
else:
  features["fun_fact"] = "Not Found"

family_match = re.search(r"(live with my family|there are .*? people in my family|my family)", text, re.IGNORECASE)
if family_match:
        features["family_info"] = family_match.group(0)
        Score+=2
else:
  features["family_info"] = "Not Found"

    # Origin / Location
origin_match = re.search(r"(i am from|my parents are from|i live in)\s+([A-Za-z\s]+)", text, re.IGNORECASE)
if origin_match:
        features["origin"] = origin_match.group(2).strip()
        Score+=2
else:
  features["origin"] = "Not Found"

    # Ambition / Goal
ambition_match = re.search(r"(i want to become|my dream is to|my ambition is to)\s+(.*?)(\.|$)", text, re.IGNORECASE)
if ambition_match:
        features["ambition"] = ambition_match.group(2).strip()
        Score+=2
else:
  features["ambition"] = "Not Found"

    # Strength or Achievement
strength_match = re.search(r"(i am good at|my strength is|i achieved|i have won)\s+(.*?)(\.|$)", text, re.IGNORECASE)
if strength_match:
        features["strengths"] = strength_match.group(2).strip()
        features["achievements"] = strength_match.group(2).strip()
        Score+=2
else:
  features["strengths"] = "Not Found"
  features["achievements"] = "Not Found"

features
```

```python
# The 'text_lower' variable already contains the original text in lowercase string format from an earlier step.
# We can directly use its word count.
# The line 'text_lower=text.lower()' was causing an error because 'text' itself had become a list.
# Displaying the word count of the already existing 'text_lower' string.
print(f"Word count of text_lower: {len(text_lower.split())}")
```

```python
import language_tool_python

def grammar_score(text_content):
    tool = language_tool_python.LanguageTool('en-US')
    errors = tool.check(text_content)
    total_words = len(text_content.split())

    score = max(0, 100 - (len(errors) / total_words * 100))

    return round(score, 2), errors

score, mistakes = grammar_score(text_lower)
Grammar_score=(len(mistakes)/len(text_lower)*100/10)
print("Grammar Score:", Grammar_score)
print("Mistakes Found:", len(mistakes))
```

```python
if Grammar_score>0.9:
  Score+=10
elif Grammar_score<0.9 and Grammar_score>0.7:
  Score+=8
elif Grammar_score<0.7 and Grammar_score>0.5:
  Score+=6
elif Grammar_score<0.5 and Grammar_score>0.3:
  Score+=4
elif Grammar_score<0.3:
  Score+=2
```

```python
import re

def vocabulary_richness(text):
    words = re.findall(r'\b\w+\b', text.lower())

    total_words = len(words)
    unique_words = len(set(words))

    TTR = unique_words / total_words if total_words else 0
    ttr=round(TTR, 3)
    return {
        "Total Words": total_words,
        "Unique Words": unique_words,
        "Vocabulary Richness (TTR)": ttr
    }

text = text_lower
print(vocabulary_richness(text))
```

```python
result_dict = vocabulary_richness(text)
total_words = result_dict["Total Words"]
unique_words = result_dict["Unique Words"]
ttr = result_dict["Vocabulary Richness (TTR)"]
```

```python
if ttr > 0.9 and ttr<1:
  Score += 10
elif ttr > 0.7 and ttr<0.9:
  Score += 8
elif ttr > 0.5 and ttr<0.7:
  Score += 6
elif ttr > 0.3 and ttr<0.5:
  Score += 4
else:
  Score += 2
```

```python
import re

def filler_word_rate(text):
    filler_words = [
        "um", "uh", "like", "you know", "actually", "basically",
        "literally", "kinda", "sort of", "well", "so", "i mean",
        "okay", "right", "hmm"
    ]

    text_clean = text.lower()


    total_words = len(re.findall(r'\b\w+\b', text_clean))


    filler_count = 0
    for filler in filler_words:
        filler_count += len(re.findall(rf"\b{re.escape(filler)}\b", text_clean))

    rate = (filler_count / total_words * 100) if total_words > 0 else 0
    rate=round(rate, 2)
    return {
        "Total Words": total_words,
        "Filler Count": filler_count,
        "Filler Word Rate (%)": round(rate, 2)
    }
```

```python
filler_word_rate(text_lower)
```

```python
result_dict = filler_word_rate(text_lower)
total_words = result_dict["Total Words"]
filler_count = result_dict["Filler Count"]
rate = result_dict["Filler Word Rate (%)"]
```

```python
if rate==0 and rate<3:
  Score+=15
elif rate>4 and rate<6:
  Score+=12
elif rate>7 and rate<9:
  Score+=9
elif rate>10 and rate<12:
  Score+=16
elif rate>=13:
  Score+=3
```
